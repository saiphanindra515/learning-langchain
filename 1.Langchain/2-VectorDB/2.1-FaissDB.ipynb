```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter


doc_loader = TextLoader("speech.txt")
documents = doc_loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=30)
docs = text_splitter.split_documents(documents)
```

```python
embeddings = OllamaEmbeddings()
db = FAISS.from_documents(docs, embeddings)
db
```

```python
query_ = "held annually"
docs_ = db.similarity_search(query_)
docs_ 
```

```python
retriever = db.as_retriever()
d_ = retriever.invoke(query_)
d_
```

```python
docs_and_score = db.similarity_search_with_score(query_)
docs_and_score
```

```python
embedding_vector = embeddings.embed_query(query_) 
sm_by_vector = db.similarity_search_by_vector(embedding_vector)
sm_by_vector
```

```python
db.save_local("faiss_index")
```

```python
new_db = FAISS.load_local("faiss_index", embeddings, allow_dangerous_local_deserialization=True)
```
